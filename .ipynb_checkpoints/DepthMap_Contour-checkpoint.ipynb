```python
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
% matplotlib inline
```

```python
#CSV with columns: Longitude, Latitude, and Depth only. Columns with blanks are cleared out previously
dive = pd.read_csv("/Users/lornasong/pydir/sh-data-format/contour_data.csv",
                   index_col = None,  header = 0,  low_memory = False,  na_values = '')
```

```python
#Grab x, y coordinates and z value. Make as array
dive = dive.drop_duplicates(['LONGITUDE', 'LATITUDE'])
x = dive['LONGITUDE'].as_matrix()
y = dive['LATITUDE'].as_matrix()
z = dive['DEPTH'].as_matrix()
```

```python
#Transform x, y, z into grid
xi = np.linspace(min(x),max(x),583)
yi = np.linspace(min(y),max(y),583)
zi = griddata((x, y), z, (xi[None,:], yi[:,None]), method='cubic')
```

```python
v = np.linspace(3, 31, 29, endpoint=True)#Sets color bar

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10, 10))#Figure size

"""Plot Map"""
contour = plt.contour(xi,yi,zi,v,linewidths=0.5,colors='k')#Contour lines
contour = plt.contourf(xi,yi,zi,v, cmap=plt.cm.jet_r)#Contour colors
#plt.scatter(x,y,marker='o',c='b',s=5) - plots data points

"""Legend"""
plt.colorbar(ticks = v)#Draw colorbar

"""Clean up plot area"""
ax.grid(False)
ax.yaxis.set_ticks_position('none')
ax.xaxis.set_ticks_position('none')
ax.set_frame_on(False)

"""Set titles and labels"""
ax.set_title('Depth Contour Map of Corral', fontsize = 20, alpha = 0.7, ha = 'left')
ax.title.set_position((0,1.05))
ax.set_xlabel("Longitude", fontsize = 14)
ax.set_ylabel("Latitude", fontsize = 14)

plt.show()
```
