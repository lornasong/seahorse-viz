```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dive = pd.read_csv("/Users/lornasong/pydir/sh-data-format/dive_data_12.7.csv",
                   index_col = None,  header = 0,  low_memory = False,  na_values = '')
seahorse = pd.read_csv("/Users/lornasong/pydir/sh-data-format/seahorse_data_12.1.csv",
                       index_col = None,  header = 0,  low_memory = False,  na_values = '')
dive.dtypes
```

```python
#Create MERGE dataframe of dive and seahorse data
merge = pd.merge(dive,  seahorse,  how = "left",  on = ["SITE",  "DATE"])
```

```python
#From MERGE create ENVIRON dataframe by melting
environment = ['PENCIL_URCHIN', 'HYDROZOAN', 'SOFT_CORAL', 'ANEMONE', 'SPONGE_SOL', 'SPONGE_SUB', 'SEAPEN', 'SHELL']

df = merge.groupby("SITE").mean().reset_index() #Find average count of each environment per site
df.fillna(0, inplace = True)#Replace NaN with zeroes

environ = pd.melt(df, id_vars = ['DATE',  'SITE', 'LONGITUDE',  'LATITUDE', 'DEPTH_x', 'SST', 'BT', 'VISIBILITY', 'SEAHORSES'],
                  value_vars = environment)
```

```python
#From MERGE create SH dataframe by clearing out null values
sh = merge.dropna(subset=['SEX [m/f]']) 
sh['SEX [m/f]'].replace('F', 'Female', inplace = True)
sh['SEX [m/f]'].replace('M', 'Male', inplace = True)
sh['SEX [m/f]'].replace('J', 'Juvenile', inplace = True)
```

```python
#Declare lists to prepare for plot
rows = len(environment)
trellis = environment
gender = ['Juvenile', 'Female', 'Male']
#colors = ['#006666', '#FFCC66', '#6B6B6B']
#colors = ['#006666', '#321A36', '#BE4F3F']
colors = ['#006666', '#FFCC66', '#321A36']
titles = ['Pencil Urchin', 'Hydrozoan', 'Soft Coral', 'Anemone', 'Sponge (Solitary)',
          'Sponge (Substrate)', 'Seapen', 'Shell']
```

```python
t = 0#Loops through titles
fig, axes = plt.subplots(nrows = rows, ncols = 1, figsize = (10, 115))

for x, i in enumerate(trellis):
#x variable loops through each subplot
#i variable loops through each plot in trellis
    
    """Plot environmental observations (varies by subplot)"""
    pared = environ[environ['variable']==i]#Filters environ by environment factor
    axes[x].scatter(pared['LONGITUDE'], pared['LATITUDE'], s = pared['value']*4,
                    edgecolor = 'none', alpha = 0.3)
        
    """Environment size-legend"""
    quantile = pared['value'].quantile([0.25, 0.5, 0.75, 1])#Gets number at the quartiles
    labels = quantile.apply(np.round)
    labels = labels.astype(int)
    l1 = plt.scatter([],[], s=labels[0.25]*4, edgecolors='none', alpha = 0.3)#Draws dots for legend
    l2 = plt.scatter([],[], s=labels[0.5]*4, edgecolors='none', alpha = 0.3)
    l3 = plt.scatter([],[], s=labels[0.75]*4, edgecolors='none', alpha = 0.3)
    l4 = plt.scatter([],[], s=labels[1]*4, edgecolors='none', alpha = 0.3)
    
    
    legend1 = axes[x].legend([l1, l2, l3, l4], labels, ncol=4, frameon=False, fontsize= 12, 
                bbox_to_anchor=[1, 0.11], handlelength= 2, handletextpad=1, columnspacing=1,
                title='Size of Average Count', scatterpoints = 1)
    
    plt.setp(legend1.get_title(), fontsize=14, alpha=0.7)
    legend1.get_title().set_position((0, 10))
    plt.gca().add_artist(legend1)#Manually adds legend1 or else it will be replaced by legend2
    
    """Plot seahorses (same for all subplots). Loops through each demographic to plot color"""
    col = 0
    jitter_long = 0.0005
    jitter_lat = 0.0005
    for s in gender:
        
        sex = sh[sh['SEX [m/f]'] == s]
        longitude = [h + np.random.uniform(-jitter_long, jitter_long) for h in sex['LONGITUDE']]
        latitude = [h + np.random.uniform(-jitter_lat, jitter_lat) for h in sex['LATITUDE']]
        
        axes[x].scatter(longitude, latitude, marker = 'x',
                    c = colors[col], edgecolor= colors[col], s = 40, linewidth = 4,
                    label = s.format(), alpha = 0.9)
        col += 1
    
    """Seahorse legend"""
    legend2 = axes[x].legend(bbox_to_anchor=(1.05, 1), scatterpoints = 1, borderpad = 0.5, #bbox moves the entirety of legend
                             frameon = False, title = 'Demographic of\nSeahorse Observation:')
    
    plt.setp(legend2.get_title(), fontsize= 14, alpha = 0.9, ha = 'left')
    legend2.get_title().set_position((-25,0)) #moves title with respect to legend text
    
    """Clean up plot"""
    axes[x].grid(False)
    axes[x].yaxis.set_ticks_position('none')
    axes[x].xaxis.set_ticks_position('none')
    #axes[x].set_frame_on(False)
    
    """Set titles and labels"""
    axes[x].set_title("Map of Average Count of " + titles[t] + " Observations", fontsize=20,
                      alpha = 0.7, ha = 'left')
    axes[x].title.set_position((0,1.05))
    t += 1 #Loops through title list
    
    axes[x].set_xlabel("Longitude", fontsize = 14)
    axes[x].set_ylabel("Latitude", fontsize = 14)
```
