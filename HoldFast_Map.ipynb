```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
seahorse = pd.read_csv("/Users/lornasong/pydir/sh-data-format/seahorse_data_12.22.csv",
                       index_col = None,  header = 0,  low_memory = False,  na_values = '')
dive = pd.read_csv("/Users/lornasong/pydir/sh-data-format/dive_data_12.7.csv",
                   index_col = None,  header = 0,  low_memory = False,  na_values = '')
merge = pd.merge(dive,  seahorse,  how = "left",  on = ["SITE",  "DATE"])
```

```python
habitat = pd.DataFrame()
habitat["LONG"] = merge["LONGITUDE"]
habitat["LAT"] = merge["LATITUDE"]
habitat["HABITAT"] = merge["HABITAT"]
habitat["SPECIES"] = merge["SPECIES"]
habitat["SNOUT"] = merge["SNOUT [cm]"]
habitat["HEAD"] = merge["HEAD [cm]"]
habitat["TRUNK"] = merge["TRUNK [cm]"]
habitat = habitat.dropna(subset=['HABITAT'])
```

```python
#Grab Unique Holdfasts in a list
holds = pd.Series(habitat['HABITAT'].values).unique()

print len(holds)
print holds

#Set colors for each holdfasts
colors = ["#030422", "#2E071A", "#80272B", "#F98842", "#F7D57E", "#1D444B", "#3E909F", "#C96B61"]
```

```python
#Jitter longitude and latitude so that points don't overlap
jitter_long = 0.0005
jitter_lat = 0.0005

habitat['j_LONG'] = [h + np.random.uniform(-jitter_lat, jitter_lat) for h in habitat['LAT']]
habitat['j_LAT'] = [h + np.random.uniform(-jitter_long, jitter_long) for h in habitat['LONG']]
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (10, 10))#Figure size
col = 0;

"""Plot holdfast points by color"""
for h in holds:
    temp = habitat[habitat["HABITAT"] == h]
    
    ax.scatter(temp["j_LONG"], temp["j_LAT"], marker = 'x', c = colors[col], edgecolor = colors[col],
               s = 40, linewidth = 4, label = h.format(), alpha = 0.9)
    col += 1

"""Set Legend"""
legend = ax.legend(bbox_to_anchor=(1.05, 1), scatterpoints = 1, borderpad = 0.5, #bbox moves the entirety of legend
                            frameon = False, title = 'Habitat of\nSeahorse Observation:')
#plt.setp(legend.get_title(), fontsize= 14, alpha = 0.9, ha = 'left')
#legend.get_title().set_position((-25,0)) #moves title with respect to legend text

"""Set titles and labels"""
ax.set_title('Observed Seahorse Habitat', fontsize = 20, alpha = 0.7, ha = 'left')
ax.title.set_position((0,1.05))
ax.set_xlabel("Longitude", fontsize = 14)
ax.set_ylabel("Latitude", fontsize = 14)

plt.show()
```
